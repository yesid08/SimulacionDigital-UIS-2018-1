```python
import numpy as np
import math
import pandas as pd
PLL=0 ;MU = 1;A = 2 ; B=4 ;AP =4 ; BP =6
PE1=0  ; PE2 = 0
TQ = 0 ; TQ2 = 0
PS1 = (np.random.random()*(B-A) + A) ; PS2 = (np.random.random()*(B-A) + A) ;  
TLL = 0 ; TLL2 = 0
TA = 0 ; TA2 = 0 ; TAC2 = 0
TE1 = 0 ; TE2 = 0 ; TEC2 = 0
NQ= 0 ; NQ2 = 0
MQ= 0 ; MQ2 = 0
TN = 0
ENTRADASC1 = [] ; LLEGADASC1 = [] ; SALIDASC1 = []
ENTRADASC2 = [] ; LLEGADASC2 = [] ; SALIDASC2 = []
PS = (np.random.random()*(BP-AP) + AP) ; PE = PS
TABLA1 = pd.DataFrame(columns=['TN' , 'PLL', 'PE1', 'PE2', 'PS1' , 'PS2' , 'ACCION','CLIENTE'])
TABLA2 = pd.DataFrame(columns=['TN' ,'PE' , 'PS','ACCION','CLIENTE'])
pd.set_option('display.max_rows', None )
pd.set_option('display.max_columns', 100)
while (TN < MU*60):
    if (NQ2>0):
        TN = min(PLL,PE1,PE2,PS1,PS2,PE,PS)
    else:
        TN = min(PLL,PE1,PE2,PS1,PS2)
    
    #UN CLIENTE ENTRA A LA COLA 2    
    if( TN == PE):
        TEC2 = TEC2 + 1
        ENTRADASC2.append(TN)
        TABLA2.loc[len(TABLA2)] = [TN,PE,PS,'Entrada a la cola',TEC2]
        NQ2 = NQ2 - 1
        PE = PS
        
    #UN CLIENTE ES ATENDIDO EN LA COLA 2
    if(TN == PS ):
        TAC2 = TAC2 + 1
        TABLA2.loc[len(TABLA2)] = [TN,PE,PS,'Salida de la cola',TAC2]
        SALIDASC2.append(TN)
        PS = PE+(np.random.random()*(BP-AP) + AP)
        PE = PS
        
    #LLEGA UN CLIENTE A LA COLA
    if (TN == PLL):
        TLL= TLL+1
        TABLA1.loc[len(TABLA1)] = [TN, PLL,PE1 ,PE2, PS1,PS2 ,'LLEGADA A LA COLA',TLL]
        LLEGADASC1.append(TN)
        TQ = TQ + NQ
        MQ = max(MQ,NQ)
        NQ = NQ + 1
        PLL= PLL + (-MU*np.log(np.random.random()))    
    
    #UNO DE LOS CAJEROS ATIENDE AL CLIENTE
    if (TN == PS1 or TN == PS2):
        #UN CLIENTE FUE ATENDIDO POR EL CAJERO #1
        if(TN ==PS1):
            TA = TA+1
            TABLA1.loc[len(TABLA1)] = [TN, PLL,PE1 ,PE2, PS1,PS2 ,'SALIDA DEL CAJERO 1',TA]
            SALIDASC1.append(TN)
            PS1 = PE1  +  (np.random.random()*(B-A) + A)
        #UN CLIENTE FUE ATENDIDO POR EL CAJERO #2
        else:
            TA = TA+1
            TABLA1.loc[len(TABLA1)] = [TN, PLL,PE1 ,PE2, PS1,PS2 ,'SALIDA DEL CAJERO 2',TA]
            SALIDASC1.append(TN)
            PS2 = PE2  +  (np.random.random()*(B-A) + A)
            
        #EL 60% DE LOS ATENDIDOS ENTRA A LA FILA 2
        if(0.2<=np.random.random()<=0.8):
            TLL2 = TLL2 +1
            if(TEC2==0):
                PE = TN
                PS = TN + (np.random.random()*(BP-AP) + AP)
            else:
                PE = PS
            TABLA2.loc[len(TABLA2)] = [TN,PE,PS,'LLegada de la cola',TLL2]
            LLEGADASC2.append(TN)
            NQ2 = NQ2 + 1
            TQ2= TQ2 + NQ2
            MQ2 = max(MQ2,NQ2)
             
    #EL CLIENTE ENTRA A UNO DE LOS DOS CAJEROS
    if (TN == PE1 or TN == PE2):
        #Verificamos que haya gente en la cola
        if (NQ > 0):
        
            #los dos cajeros tienen los mismos tiempos de entrada, elegimos uno al azar
            if(PE1 == PE2):
                #ATIENDE EL 1
                if(np.random.random()<0.5):
                    TE1 = TE1 +1
                    TABLA1.loc[len(TABLA1)] = [TN, PLL,PE1 ,PE2, PS1,PS2 ,'ENTRADA ALEATORIA AL CAJERO 1',TE1]
                    ENTRADASC1.append(TN)
                    NQ = NQ-1
                    PE1 = PS1
                    if (NQ == 0  and  PLL > PS1):
                        PE1 = PLL 

                #ATIENDE EL 2
                else:
                    TE2 = TE2 +1
                    TABLA1.loc[len(TABLA1)] = [TN, PLL,PE1 ,PE2, PS1,PS2 ,'ENTRADA ALEATORIA AL CAJERO 2',TE2]
                    ENTRADASC1.append(TN)
                    NQ = NQ-1
                    PE2 = PS2
                    if (NQ == 0  and  PLL > PS2):
                        PE2 = PLL
                        
            #Los tiempos entre cajeros son diferentes y se elige el mejor       
            else:
                #EL CLIENTE ENTRA AL #1
                if(TN == PE1):
                    TE1 = TE1 +1
                    TABLA1.loc[len(TABLA1)] = [TN, PLL, PE1, PE2,  PS1, PS2, 'ENTRADA AL CAJERO 1',TE1]
                    ENTRADASC1.append(TN)
                    NQ = NQ-1
                    PE1 = PS1
                    if (NQ == 0  and  PLL > PS1):
                        PE1 = PLL 
                #EL CLIENTE ENTRA AL #2    
                else:
                    TE2 = TE2 +1
                    TABLA1.loc[len(TABLA1)] = [TN, PLL, PE1, PE2,  PS1, PS2, 'ENTRADA AL CAJERO 2',TE2]
                    ENTRADASC1.append(TN)
                    NQ = NQ-1
                    PE2 = PS2
                    if (NQ == 0  and  PLL > PS2):
                        PE2 = PLL
            
        else:
            #LLEGA UN CLIENTE A LA COLA
            TLL= TLL+1
            TABLA1.loc[len(TABLA1)] = [TN, PLL, PE1, PE2,  PS1, PS2, 'NO HAY CLIENTES, ESPERAMOS QUE LLEGUE UNO EN LA COLA',TLL]
            LLEGADASC1.append(TN)
            TQ = TQ + NQ
            MQ = max(MQ,NQ)
            NQ = NQ + 1
            PLL= PLL + (-MU*np.log(np.random.random()))
    
shape = min(len(LLEGADASC1),len(ENTRADASC1), len (SALIDASC1))
shape2 = min(len(LLEGADASC2),len(ENTRADASC2), len (SALIDASC2))
min1 = 9999
min2 = 9999
max1 = 0
max2 = 0
promedio1 = 0
promedio2 = 0
print ("COLA 1")
TABLA1

for i in range (0,shape):
    print ("El tiempo que demoró el cliente "+str(i+1)+ " fue de :", SALIDASC1[i] -LLEGADASC1[i])
    promedio1 = promedio1 + (SALIDASC1[i] -LLEGADASC1[i])
    min1 = min(min1,SALIDASC1[i] -LLEGADASC1[i])
    max1 = max(max1,SALIDASC1[i] -LLEGADASC1[i])
print ("El tiempo promedio de permanencia en la cola 1 es: ", float(promedio1)/shape)
print ("El tiempo máximo de permanencia en la cola 1 es:" , max1)
print ("El tiempo mínimo de permanencia en la cola 1 es:" , min1)
print ("Promedio de la cola 1:" , float(TQ)/float(TLL))
print ("Total de atendidos cola 1: ", TA)
print ("Llegaron a la cola 1:  " , TLL)
print ("Máximo tamaño de la cola 1:" , MQ)

print ("COLA 2")
for i in range (0,shape2):
    print ("El tiempo que demoró el cliente "+str(i+1)+ " fue de :", SALIDASC2[i] -LLEGADASC2[i] )
    promedio2 = promedio2 + (SALIDASC2[i] -LLEGADASC2[i])
    min2 = min(min2,SALIDASC2[i] -LLEGADASC2[i])
    max2 = max(max2,SALIDASC2[i] -LLEGADASC2[i])
print ("El tiempo promedio de permanencia en la cola 2 es :" , float (promedio2)/shape2)
print ("El tiempo máximo de permanencia en la cola 2 es:" , max2 )
print ("El tiempo mínimo de permanencia en la cola 2 es:" , min2)
print ("Promedio de la cola 2:" , float(TQ2)/float(TLL2))
print ("Total de atendidos cola 2: ", TAC2)
print ("Llegaron a la cola 2:  " , TLL2)
print ("Máximo tamaño de la cola 2:" , MQ2)
```
