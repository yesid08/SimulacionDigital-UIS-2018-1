# I) condiciones iguales a cero, los incisos a,b y c se distinguen por la variación del voltaje de entrada.

## Inciso a) 
las condiciones iniciales son:  
+ cero $X1(0) = X2(0) = 0$  
+ el voltaje de entrada es: una función impulso,
+ Para $V(0) = 1000$ , $V = 0$ para cualquier otro t

```python

import numpy as np
import pandas as pd
import sympy as sp
import matplotlib.pyplot as plt
N = 1*10**(3) +200
V = np.zeros(N)
Columns = ['Tiempo','Voltaje' , 'X2' , 'X1' ]
Tabla = pd.DataFrame(columns=Columns)
R = 40
L = 0.8
C = 40*10**(-6)
T = 1.*10**(-3)
X1 = np.zeros(N)
X2 = np.zeros(N)
V[0] = 1000
Tiempo = np.zeros(N)
for k in range (N-1):
    Tiempo[k] = k*T
    X1[k+1] = T*( X2[k] - (1.*R/L)*X1[k] ) + X1[k]
    X2[k+1] = ( (1./L)*V[k] - (1./(L*C))*X1[k] )*T + X2[k]
    Tabla.loc[len(Tabla)] = [ Tiempo[k], V[k], X2[k] , X1[k]]

plt.xlabel('Tiempo en [ms]')
plt.ylabel('Carga Q en [c]')
plt.xscale('linear')
plt.grid()
plt.title("Carga del circuito RLC")
plt.plot(X1, 'r',linewidth=2)
plt.legend("Q")
plt.show()
plt.legend("V")
plt.title("Voltaje aplicado al circuito RLC")
plt.xlabel('Tiempo en [ms]')
plt.ylabel('Voltaje en [v]')
plt.xscale('linear')
plt.grid()
plt.plot(V,'r',linewidth=2)
plt.legend("V")
plt.show()
Tabla
```

## inciso b)
las condiciones iniciales son:  
+ cero $X1(0) = X2(0) = 0$  
+ el voltaje de entrada es la función $e^{-3t}*cos(60t)$

```python
import numpy as np
import pandas as pd
import sympy as sp
import matplotlib.pyplot as plt
N = 1*10**(3) + 200
V = np.zeros(N)
Columns = ['Tiempo','Voltaje' , 'X2' , 'X1' ]
Tabla = pd.DataFrame(columns=Columns)
R = 40
L = 0.8
C = 40*10**(-6)
T = 1.*10**(-3)
X1 = np.zeros(N)
X2 = np.zeros(N)
Tiempo = np.zeros(N)
for k in range (N-1):
    Tiempo[k] = k*T
    V[k] = sp.exp(-3*Tiempo[k])*sp.cos(60*Tiempo[k])
    X1[k+1] = T*( X2[k] - (1.*R/L)*X1[k] ) + X1[k]
    X2[k+1] = ( (1./L)*V[k] - (1./(L*C))*X1[k] )*T + X2[k]
    Tabla.loc[len(Tabla)] = [ Tiempo[k], V[k], X2[k] , X1[k]]
plt.xlabel('Tiempo en [ms]')
plt.ylabel('Carga Q en [c]')
plt.xscale('linear')
plt.grid()
plt.plot(X1, 'b',linewidth=2)
plt.legend("Q")
plt.title("Carga del circuito RLC")
plt.show()
plt.xlabel('Tiempo en [ms]')
plt.ylabel('Voltaje V en [v]')
plt.xscale('linear')
plt.grid()
plt.plot(V, 'b',linewidth=2)
plt.legend("V")
plt.title("Voltaje aplicado al circuito RLC")
plt.show()
Tabla
```
