## 2.1. GENERADORES CONGRUENCIALES LINEALES
$N_{i+1} = A*N_i + C \pmod M$
### Ejemplo:
  
Calcular los primeros dos números generados según la fórmula dada y
según los siguientes valores:  
+ $N_0 = 17$ 
+ $A = 10$
+ $M = 30$
+ $C = 0$  
página 43


```python
#Funcion que me genera los 2 primeros números aleatorios por medio de un generador congruencial lineal.
def generadorCongruencialLineal2Primeros(N0,A,M,C):
    N = N0
    data= []
    for i in range (2):
        data.append([i,N])
        N = (A*N + C) % M
    tabla = pd.DataFrame(columns=["n" , "Xn"] , data = data)
    display(tabla)
```

```python
#Ejecución del programa
N0 = 17
A = 10
M = 30
C = 0
generadorCongruencialLineal2Primeros(N0,A,M,C)
```

## 2.1.1. Método aditivo de congruencias:
$N_{i+1} = [N_i + N_{i-k}] \pmod M$
### Ejemplo 
Sea una secuencia de enteros dados: x1, x2, x3, x4 y x5 (57, 34,
89, 92 y 16), por tanto k =5. Consideremos m = 100. La secuencia
anterior puede ser ampliada por este método:  
página 44


```python
#funcion que genera M numeros aleatorios por el método aditivo de congruencias.
def aleatoriosMetodoAditivoDeCongruencias(X , M):
    K = len(X)
    print ("K = ",K)
    print ("X aleatorios generados")
    for i in range(K,M+1):
        X.append((X[i - 1]+ X[i - K]) % M)
    tablaX = pd.DataFrame(columns=["Xi"] , data = X , index=range(1,M+2))
    display(tablaX)
    Xnorm = np.array(X)
    Xnorm = Xnorm/M
    tablaXnorm  = pd.DataFrame(columns=["XiNorm"] , data = Xnorm , index = range(1,M+2))
    print ("X aleatorios normalizados generados")
    display(tablaXnorm)
```

```python
#Ejecución del programa
M = 100
X = []
X.append(57)
X.append(34)
X.append(89)
X.append(92)
X.append(16)
aleatoriosMetodoAditivoDeCongruencias(X, M)
```

## 2.1.2. Método multiplicativo de congruencias:
$X_{i+1} = A* X_i \pmod M$
### Ejemplo:
Tomando :  
    $A = 181$  
    $Xo = 501$  
    $M = 32768$  
página 47  

```python
#funcion que permite generar 10 aleatorios
def funcionGenerar10Aleatorios(A,X0,M):
    data= []
    columns = ["n" , "Xn"]
    X = X0
    #Desde 1 hasta 10, python no toma el último valor
    for i in range ( 1 , 10 + 1 ):
        X = X * ( A %  M  )
        X = X - (X//M) * M
        #Si el resultado es negativo se utiliza:
        #Y = X/M
        data.append([i,X])
    tabla = pd.DataFrame(columns=columns , data=data)
    display(tabla)
```

```python
#Ejecución del programa
A = 181
X0 = 501
M = 32768
funcionGenerar10Aleatorios(A = A , X0 = X0 , M = M)
```

```python
#Función que permite comparar el periodo experimental con el esperado
def funcionPeridoEsperado(b,A,X):
    import pandas as pd
    data= []
    columns = ["n" , "Xn"]
    M = 2**(b)
    print ("M = ",M)
    H = 0
    while ( True ):
        X = A * (X % M)
        X = X - (X//M) * M
        if(H == 0):
            X0 = X
        if( X == X0 and H != 0 ):
            print("El periodo experimental de los números es : ",H)
            break;
        H = H + 1
        data.append([H,X])
    print ("El periodo esperado de los números es :", M//4)
    tabla = pd.DataFrame(columns=columns , data=data)
    display (tabla)
```

## 2.1.3 Método mixto de congruencias
$X_{i+1} = [A * X_i + C] \pmod M$
### Ejercicio 1
Hallar A y Xo tal que garanticen el máximo período, con un M = 32.  (Método Multiplicativo).   
página 54

```python
#funcion que permite garantizar el máximo periodo en el método multiplicativo
def metodoMultiplicativoPeriodoMaximo(M,X0):
    columnas = ["n","Xn"]
    data= []
    A = int(math.sqrt(M))
    H = 0
    X = X0
    while((X != X0) or (H == 0) ):
        data.append([H,X])
        X = A * X % M
        H = H + 1
    tabla = pd.DataFrame(columns=columnas,data=data)
    display(tabla)
    return H
```

```python
#Ejecución del programa 1
M = 32
X0 = 11
H = metodoMultiplicativoPeriodoMaximo(M,X0)
print("El periodo máximo esperado es: ",M//4)
print("El periodo máximo experimental es: ",H)
```

```python
#Ejecución del programa 2
M = 32
X0 = 7
H = metodoMultiplicativoPeriodoMaximo(M,X0)
print("El periodo máximo esperado es: ",M//4)
print("El periodo máximo experimental es: ",H)
```

## Método aditivo de congruencias
### Ejercicio 2
Para la siguiente ecuación de recurrencia (Método Aditivo de Congruencias)  
$X_n = [X_{n-1} + X_{n-2}] \pmod {16} $  
Hacer un programa para hallar el período.  
página 55

```python
#Función que permite calcular el periodo maximo por el método aditivo de congruencias.
def MetodoAditivoCongruenciasPeriodoMaximo(S1,S2,M):
    import pandas as pd
    data= []
    columns = ["n" , "Xn"]
    U = S2  
    P = S1
    H = 0
    while (True): 
        N= (U + P) % M
        H = H + 1
        data.append([H,N])
        if N == S2 and U == S1:
            break;
        P = U 
        U = N
    tabla = pd.DataFrame(columns=columns,data=data)
    return H,tabla
```

```python
#Ejecución del programa
S1 = 1
S2 = 8
M = 16
periodo, tabla = MetodoAditivoCongruenciasPeriodoMaximo(S1 = S1 , S2 = S2, M = M)
print ("El periodo obtenido es: ",periodo)
display(tabla)
```
